## getaddrinfo 구조 살펴보기
- 그림은 알겠는데 실제로 어떤 데이터가 들어있는지,
- 클라이언트에서의 addrinfo 구조체와 서버에서의 addrinfo가 어떤 정보들을 담고 있는지 알아보자

```python
# 서버입장의 getaddrinfo
HOST = None
PORT = 8000
# 어떤 host의 요청이든 8000포트로 들어오면 받아주는 소켓 객체를 반환(C였으면 file descriptor를 반환)
# 결과물은 IPv4를 위한 소켓주소와 IPv6를 위한 소켓주소가 하나씩 반환
socket.getaddrinfo(None, 8000, 
                   family=socket.AF_UNSPEC, # IPv4, IPv6 둘 다 
                   type=socket.SOCK_STREAM, # stream타입의 소켓
                   proto=socket.IPPROTO_TCP, # TCP 프로토콜
                   flags=socket.AI_PASSIVE, # 어떤 host의 요청이든 받아주마
                   )
```

```python
# 클라이언트 입장의 getaddrinfo

# 서버도 내 주소이기 때문에 HOST에 내 주소가 들어가지만, 실제로는 naver.com 같은 원격 서버의 주소가 들어갈 것이다.
HOST = socket.gethostbyname(socket.gethostname())
PORT = 50007
socket.getaddrinfo(HOST, PORT, socket.AF_UNSPEC, socket.SOCK_STREAM)

# 이렇게 원격 호스트의 소켓주소를 알아온 다음, 내 소켓을 만들어 connect를 한다.
# 소켓 인스턴스 생성 -> connect 시도 -> close
```
